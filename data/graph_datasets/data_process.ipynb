```python
# process raw graph data and output files as required. we will convert raws data into
# an dict object, where the key is node id and the value is {'id':<id>, 'feature': <feature>, 'label': <label>, 'edge': [<id, node, w>...<id, node, w>]}
```

```python
import pandas as pd
import numpy as np
from tqdm import tqdm
```

```python
net_path= '/home/yf302/Desktop/Kai/graph_datasets/citeseer/citeseer.content'
edge_path = '/home/yf302/Desktop/Kai/graph_datasets/citeseer/citeseer.cites'
```

```python
graph = pd.read_table(net_path, header=None)
graph[0] = graph[0].astype("string")
```

```python
graph.head()
```

first column represents node id, last column represents label, middle columns represent features

```python
edge = pd.read_table(edge_path, header=None, dtype='string')
```

```python
edge.head()
```

```python
# get node list, label list
node_list = graph.iloc[:,0].unique()
class_list = graph.iloc[:, -1].unique()
```

```python
# node -> id, label -> number
node2id = {node:id for id, node in enumerate(node_list)}
class_mapping = {label:id for id, label in enumerate(class_list)}
```

```python
node2id
```

```python
class_mapping
```

```python
dataset = {}
for node_name, id in node2id.items():
    data = {}
    node = graph[graph.iloc[:,0]==node_name].values.reshape(-1)
    data['id'] = id
    data['feature'] = node[1:-1]
    label = node[-1]
    data['label'] = class_mapping[label]
    neighbors = edge[edge.iloc[:,1]==node_name].iloc[:, 0].values
    edges = []
    for x in neighbors:
        try:
            x = node2id[x]
        except:
            print('Key Error', x)
            continue
        edges.append([id, x, 1])
    data['edge'] = edges
    dataset[node_name] = data
```

```python
label = []
for node_name, data in dataset.items():
    label.append([data['id'], data['label']])
label
```

```python
label_df = pd.DataFrame(label)
label_df.to_csv('citeseer/label.txt',  header=None, index=None, sep='\t')
```

```python
net = []
for node_name, data in dataset.items():
    edge = data['edge']
    for e in edge:
        net.append(e)
net
```

```python
pd.DataFrame(net).to_csv('citeseer/net.txt', header=None, index=None, sep='\t')
```

```python
features = []
for node_name, data in dataset.items():
    id = data['id']
    feature = data['feature']
    nodes = np.where(feature!=0)[0]
    raw = [id]
    line = []
    for n in nodes:
        line.append(f'{n}:1.0')
    line = ' '.join(line)
    raw.append(line)
    features.append(raw)

features
```

```python
pd.DataFrame(features).to_csv('citeseer/feature.txt', header=None, index=None, sep='\t')
```

```python
pubmed_edge = pd.read_table('/home/yf302/Desktop/Kai/graph_datasets/Pubmed-Diabetes/data/Pubmed-Diabetes.DIRECTED.cites.tab', skiprows=[0,1], header=None)
```

```python
pubmed_edge.head()
```

```python
edges = []
for e in pubmed_edge.iloc[:].values:
    cited_id = e[1].split(':')[1]
    citing_id = e[3].split(':')[1]
    edges.append([cited_id, citing_id])

edges = np.array(edges)
```

```python
edges = pd.DataFrame(edges, columns=['cited', 'citing'])
```

```python
edges.head()
```

```python
pubmed_graph = pd.read_table('/home/yf302/Desktop/Kai/graph_datasets/Pubmed-Diabetes/data/Pubmed-Diabetes.NODE.paper.tab', skiprows=[0])
```

```python
corpus = pubmed_graph.iloc[0].index.values
```

```python
vocab = []
for item in corpus:
    if 'numeric' in item:
        word = item.split(':')[1].split('-')[-1]
        vocab.append(word)
```

```python
vocab = {word:idx for idx, word in enumerate(vocab)}
node2id = {node:idx for idx, node in enumerate(set(edges.values.flatten()))}
```

```python
vocab
```

```python
node2id
```

```python
# dataset = {'nodeid':
    # {
        # 'label':0,
        # 'feature':['wordid':'weight'],
        # 'edge':['nodeid']
    # }
# }
```

```python
dataset = {}
for item in tqdm(pubmed_graph.iloc[1:].values):
    nodeid = item[0]
    id = node2id[str(nodeid)]
    label = int(item[1].split('=')[1]) - 1
    feature = [id]
    raw = []
    for word in item[2:]:
        if type(word) == str and 'summary' not in word:
            word, weight = word.split('-')[1].split('=')
            word_id = vocab[word]
            raw.append(str(word_id)+':'+str(weight)) # add [word:weight]
    feature.append(' '.join(raw))
    
    edge = []
    neighbors = edges[edges.citing==str(nodeid)].cited
    for x in neighbors:
        edge.append([id, node2id[str(x)], 1])
    dataset[nodeid] = {}
    dataset[nodeid]['id'] = id
    dataset[nodeid]['label'] = label
    dataset[nodeid]['edge'] = edge
    dataset[nodeid]['feature'] = feature
```

```python
label_df = pd.DataFrame(label)
label_df.to_csv('pubmed/label.txt',  header=None, index=None, sep='\t')
```
